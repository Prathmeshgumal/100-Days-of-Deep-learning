```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=15)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    w1=w2=b=1
    lr = 0.1
    for i in range(1000):
        for j in range(X.shape[0]):
            z = w1*X[j][0] + w2*X[j][1] + b

            if z*y[j] < 0 :
                w1 = w1 + lr*y[j]*X[j][0]
                w2 = w2 + lr*y[j]*X[j][1]
                b = b +lr*y[j]

    return w1,w2,b
```

```python
w1,w2,b= perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)
print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_input,y_input,color = 'red',linewidth=2)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
